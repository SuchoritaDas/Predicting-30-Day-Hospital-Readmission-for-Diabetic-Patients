# Diabetes 30-Day Readmission Project (Official Implementation)

The Bronze layer stores the raw dataset exactly as received from the source, without any transformation.
Its purpose is to preserve data lineage, enable reproducibility, and allow reprocessing if business logic changes later.

```python
raw_df = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/diabetes_readmission/healthcare/diabetic_data.csv")
)
```

```python
bronze_df = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/workspace/diabetes_readmissions/healthcare/diabetic_data.csv")
)
```

```python
dbutils.fs.ls("/Volumes/workspace/diabetes_readmissions/healthcare/")
```

```python
bronze_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("diabetes_readmissions.bronze_diabetes_readmission")
# This saves the DataFrame to the correct schema and table as a Delta table. Ensure the schema exists and you have write permissions.
```
